# NEURAL NETWORK

```python
class SentimentalAnalysis:
    def __init__(self,reviews,labels,hidden_nodes=10,learning_rate=0.1):
        np.random.seed(1)
        self.preprocess_network(reviews,labels)
        self.init_network(len(self.review_vocab),hidden_nodes,1,learning_rate)
    
    def preprocess_network(self,reviews,labels):    
        review_vocab=set()
        for review in reviews:
            for word in review.split(" "):
                review_vocab.add(word)

        self.review_vocab=list(review_vocab)
    
        label_vocab=set()
        for label in labels:
            for l in label.split(" "):
                label_vocab.add(l)
            
        self.label_vocab=list(label_vocab)
        
        self.word2index={}
        for i,word in enumerate(self.review_vocab):
            self.word2index[word]=i
        
        label2index={}
        for i,label in enumerate(self.label_vocab):
            label2index[label]=i
        self.label2index=label2index 
    
    def init_network(self,input_nodes,hidden_nodes,output_nodes,learning_rate):
        self.input_nodes=input_nodes
        self.hidden_nodes=hidden_nodes
        self.output_nodes=output_nodes     
        self.learning_rate=learning_rate
        
        self.weights_input_hidden=np.zeros((self.input_nodes,self.hidden_nodes))
        self.weights_hidden_output=np.random.normal(size=(self.hidden_nodes,self.output_nodes))
        self.layer_0=np.zeros((1,input_nodes))
   
    def update_input_layer(self,review):
        
        self.layer_0=self.layer_0*0
        
        for word in review.split(" "):
            if(word2index[word]>len(self.layer_0)):
                continue
            self.layer_0[0][word2index[word]]+=1
        
    def target_label(self,label):
        if(label=="POSITIVE"):
            return 1
        else:
            return 0

    def sigmoid(self,x):
        return (1/(1+np.exp(-x)))
    
    def sigmoid_output2_dervative(self,output):
        return (output*(1-output))
    
    def train(self,training_reviews,training_labels):
        
        print ("########## TRAINING ##########")
        
        correct_so_far=0
        
        print ("Total length:"+str(len(training_reviews)))
        for i in range(len(training_reviews)):
            review=training_reviews[i]
            label=training_labels[i]
            self.update_input_layer(review)
            layer_1=np.dot(self.layer_0,self.weights_input_hidden)
            layer_2=self.sigmoid(layer_1.dot(self.weights_hidden_output))
            
            layer_2_error=layer_2-self.target_label(label)
            layer_2_delta=layer_2_error*self.sigmoid_output2_dervative(layer_2)
            
            layer_1_error=layer_2_delta.dot(self.weights_hidden_output.T)
            layer_1_delta=layer_1_error
            
            self.weights_input_hidden-=self.layer_0.T.dot(layer_1_delta)*self.learning_rate
            self.weights_hidden_output-=layer_1.T.dot(layer_2_delta)*self.learning_rate
            
            if(label=="POSITIVE" and layer_2>=0.5):
                correct_so_far+=1
            elif(label=="NEGATIVE" and layer_2<0.5):
                correct_so_far-=1
        print("Correct Number is: "+str(correct_so_far))
                
    def test(self,training_reviews,training_labels):
        
        print ("########## TESTING ##########")
        
        correct=0
        print ("Total length:"+str(len(training_reviews)))
        for i in range(len(training_reviews)):
            pred=self.run(training_reviews[i])
            if(pred==training_labels[i]):
                correct+=1
        print("Correct Number is: "+str(correct))
        
   
    def run(self, review):
        self.update_input_layer(review)
        layer_1=self.layer_0.dot(self.weights_input_hidden)
        layer_2=self.sigmoid((layer_1.dot(self.weights_hidden_output)))
        
        if(layer_2[0] >= 0.5):
            return "POSITIVE"
        else:
            return "NEGATIVE"
```

# Reducing Vocablary by removing less frequent term and most frequent terms

```python
class SentimentalAnalysis:
    def __init__(self,reviews,labels,hidden_nodes=10,learning_rate=0.1):
        np.random.seed(1)
        self.preprocess_network(reviews,labels)
        self.init_network(len(self.review_vocab),hidden_nodes,1,learning_rate)
    
    def preprocess_network(self,reviews,labels): 
        
        positive=Counter()
        negative=Counter()
        total=Counter()
        
        min_count = 10
        polarity_cutoff = 0.1
        
        for i,review in enumerate(reviews):
            for word in review.split(" "):
                if(labels[i]=="POSITIVE"):
                    positive[word]+=1
                    total[word]+=1
                else:
                    negative[word]+=1
                    total[word]+=1

        pos_neg_ratio=Counter()
        for (word,cnt) in (positive.most_common()):
            if(cnt>50):
                pos_neg_ratio[word]=positive[word]/(negative[word]+1)
                
        for word,ratio in pos_neg_ratio.most_common():
            if(ratio > 1):
                pos_neg_ratio[word] = np.log(ratio)
            else:
                pos_neg_ratio[word] = -np.log((1 / (ratio + 0.01)))
                
        
        review_vocab=set()
        for review in reviews:
            for word in review.split(" "):
                if(total[word]>min_count):
                    if(word in pos_neg_ratio.keys()):
                        if((float(pos_neg_ratio[word])>=polarity_cutoff) or (float(pos_neg_ratio[word])<=-polarity_cutoff)):
                            review_vocab.add(word)
                    else:
                        review_vocab.add(word)

        self.review_vocab=list(review_vocab)
    
        label_vocab=set()
        for label in labels:
            for l in label.split(" "):
                label_vocab.add(l)
            
        self.label_vocab=list(label_vocab)
        
        self.word2index={}
        for i,word in enumerate(self.review_vocab):
            self.word2index[word]=i
        
        label2index={}
        for i,label in enumerate(self.label_vocab):
            label2index[label]=i
        self.label2index=label2index 
    
    def init_network(self,input_nodes,hidden_nodes,output_nodes,learning_rate):
        self.input_nodes=input_nodes
        self.hidden_nodes=hidden_nodes
        self.output_nodes=output_nodes     
        self.learning_rate=learning_rate
        
        self.weights_input_hidden=np.zeros((self.input_nodes,self.hidden_nodes))
        self.weights_hidden_output=np.random.normal(size=(self.hidden_nodes,self.output_nodes))
        self.layer_1=np.zeros((1,self.hidden_nodes))
   
        
    def target_label(self,label):
        if(label=="POSITIVE"):
            return 1
        else:
            return 0

    def sigmoid(self,x):
        return (1/(1+np.exp(-x)))
    
    def sigmoid_output2_dervative(self,output):
        return (output*(1-output))
    
    
#     ########### NEW ######


    def train(self,training_reviews,training_labels):
        
        training_index_list=list()
        
        for review in training_reviews:
            indices=set()
            for word in review.split(" "):
                if(word in self.word2index.keys()):
                    indices.add(self.word2index[word])
            training_index_list.append(list(indices))
                
        
        print ("########## TRAINING ##########")
        
        correct_so_far=0
        
        print ("Total length:"+str(len(training_reviews)))
        for i in range(len(training_index_list)):
            review=training_index_list[i]
            label=training_labels[i]
            
            self.layer_1*=0
            for index in review:
                if(index>self.input_nodes):
                    continue
                self.layer_1+=self.weights_input_hidden[index]
            
    
            layer_2=self.sigmoid(self.layer_1.dot(self.weights_hidden_output))
            
            layer_2_error=layer_2-self.target_label(label)
            layer_2_delta=layer_2_error*self.sigmoid_output2_dervative(layer_2)
            
            layer_1_error=layer_2_delta.dot(self.weights_hidden_output.T)
            layer_1_delta=layer_1_error
            
            for index in review:
                self.weights_input_hidden[index]-=layer_1_delta[0]*self.learning_rate
                
            self.weights_hidden_output-=self.layer_1.T.dot(layer_2_delta)*self.learning_rate
            
            if(label=="POSITIVE" and layer_2>=0.5):
                correct_so_far+=1
            elif(label=="NEGATIVE" and layer_2<0.5):
                correct_so_far-=1
        print("Correct Number is: "+str(correct_so_far))
                
    def test(self,training_reviews,training_labels):
        
        print ("########## TESTING ##########")
        
        training_index_list=list()
        
        for review in training_reviews:
            indices=set()
            for word in review.split(" "):
                if(word in self.word2index.keys()):
                    indices.add(self.word2index[word])
            training_index_list.append(list(indices))
        
        correct=0
        print ("Total length:"+str(len(training_reviews)))
        for i in range(len(training_index_list)):
            pred=self.run(training_index_list[i])
            if(pred==training_labels[i]):
                correct+=1
        print("Correct Number is: "+str(correct))
        
   
    def run(self, review):
        
        self.layer_1*=0
        
        for index in review:
            if(index>=self.input_nodes):
                continue
            self.layer_1+=self.weights_input_hidden[index]
        

        layer_2=self.sigmoid((self.layer_1.dot(self.weights_hidden_output)))
        
        if(layer_2[0] >= 0.5):
            return "POSITIVE"
        else:
            return "NEGATIVE"
```

```python
def get_most_similar_words(focus = "horrible"):
    most_similar = Counter()

    for word in mlp_full.word2index.keys():
        most_similar[word] = np.dot(mlp_full.weights_input_hidden[mlp_full.word2index[word]],mlp_full.weights_input_hidden[mlp_full.word2index[focus]])
    
    return most_similar.most_common()
```
